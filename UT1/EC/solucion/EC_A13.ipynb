#### Ejercicio 1

Has sido asignado para crear un conjunto de clases para modelar diferentes tipos de vehiculos.<br>
- Crear una clase base `Vehicle`, con los atributos `model` y `year`.
- Crear una clase `Car` y `Motorcycle` que herede de `Vehicle`.
- Cada clase tendrá sus propios atributos y métodos.
  - `Car`
    - tendrá `num_wheels`, `num_doors`.
    - tendrá un método car_info() que imprima todos los atributos.
  - `Motorcycle`
    - tendrá `num_wheels`, `num_doors`
    - tendra un método motorcycle_info() que imprima todos los atributos.


Reglas:
- Usa la sintaxis super() para llamar a los métodos de la clase padre.

Prueba tu código:
1. Crea un objeto `Car` con los datos que quieras.
2. Crea un objeto `Motorcycle` con los datos que quiera.
3. Imprime la información a través de los métodos `car_info()` y `motorcycle_info()`.



````python
class Vehicle:
  def __init__(self, model, year):
    self.model = model
    self.year = year

class Car(Vehicle):
  def __init__(self, model, year, num_wheels, num_doors):
    super().__init__(model, year)
    self.num_wheels = num_wheels
    self.num_doors = num_doors

  def car_info(self):
    print(f"Model: {self.model}, Year: {self.year}, Num Wheels: {self.num_wheels}, Num Doors: {self.num_doors}")

class Motorcycle(Vehicle):
  def __init__(self, model, year, num_wheels, num_doors):
    super().__init__(model, year)
    self.num_wheels = num_wheels
    self.num_doors = num_doors

  def motorcycle_info(self):
    print(f"Model: {self.model}, Year: {self.year}, Num Wheels: {self.num_wheels}, Num Doors: {self.num_doors}")

# Prueba del código
car = Car("Toyota", 2022, 4, 4)
motorcycle = Motorcycle("Harley Davidson", 2021, 2, 0)

car.car_info()
motorcycle.motorcycle_info()
````

#### Ejercicio 4

Basado en el ejercicio anterior, añade el método `__str__` a la clase base `Vehicle` para que imprima la información del objeto.<br>
El método `info()` debe llamar al método `__str__` para imprimir la información.

````python
class Vehicle:
  def __init__(self, model, year, num_wheels, num_doors):
    self._model = model
    self._year = year
    self._num_wheels = num_wheels
    self._num_doors = num_doors

  def __str__(self):
    return f"Model: {self._model}, Year: {self._year}, Num Wheels: {self._num_wheels}, Num Doors: {self._num_doors}"

  def info(self):
    return self.__str__()
 

# Test code
car = Vehicle("Toyota", 2022, 4, 4)
print(car)

assert str(car) == "Model: Toyota, Year: 2022, Num Wheels: 4, Num Doors: 4"
assert car.info() == "Model: Toyota, Year: 2022, Num Wheels: 4, Num Doors: 4"
````

3. Crea un objeto `Cat` y llama al método `make_sound()`.
4. Crea un objeto `Cow` y llama al método `make_sound()`.

````python
from abc import ABC, abstractmethod

class Animal(ABC):

  def __init__(self, especie, sonido_emite):
    self.especie = especie
    self.sonido_emite = sonido_emite

  @abstractmethod
  def makesound(self):
    pass


class Dog(Animal):
  
  def __init__(self):
    super().__init__("perro", "guau")

  def makesound(self):
    print(f"El {self.especie} hace {self.sonido_emite}")


class Cat(Animal):
    
  def __init__(self):
    super().__init__("gato", "miau")

  def makesound(self):
    print(f"El {self.especie} hace {self.sonido_emite}")


class Cow(Animal):
      
  def __init__(self):
    super().__init__("vaca", "muu")

  def makesound(self):
    print(f"El {self.especie} hace {self.sonido_emite}")
  

# Intentar instanciar un objeto de la clase Animal. Try/Catch captura la excepción y muestra el error. NO VISTO AÚN EN CLASE
try:
  animal = Animal()
except TypeError as e:
  print("Error:", e)
  

# Crear objetos y llamar al método makesound()
dog = Dog()
dog.makesound()

cat = Cat()
cat.makesound()

cow = Cow()
cow.makesound()
````

  - Crea un método de clase `get_category()` que devuelva el valor del atributo de clase.

````python
class Vehicle:
  _category = 'terrestre'

  def __init__(self, model, year, num_wheels, num_doors):
    self.model = model
    self.year = year
    self.num_wheels = num_wheels
    self.num_doors = num_doors

  def getCategory():
    return Vehicle._category
  

# Probar el código	
car1 = Vehicle("Toyota", 2022, 4, 4)
print(Vehicle.getCategory())
assert(Vehicle.getCategory() == "terrestre")
````
