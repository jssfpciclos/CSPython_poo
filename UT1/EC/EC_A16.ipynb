### Ejercicio 2

- Diseña una clase llamada `Cliente` con las siguientes características:
    - Sus atributos son: nombre, dirección, teléfono y nif. 
    - Tiene los siguientes métodos:
        - Un constructor.
        - Los métodos getters de todos los atributos (tipo @property).

- Diseña una clase llamada `CuentaBancaria` que:<br>
  - Sus atributos son: 
    - titular (que es un objeto de la clase Cliente). 
    - saldo (float).
    - número de cuenta.
  - Métodos:
    - depositar(importe): Incrementa el saldo en la cantidad que se indique.
    - transferencia(importe, Cuenta): Mueve la cantidad a otra cuenta siempre que en la cuenta origen haya dinero suficiente para poder hacerla.<br>
     Devolverá True si se ha podido hacer la transferencia y False en caso contrario.


Utiliza la siguiente validación para comprobar la funcionalidad:
```python
cliente1 = Cliente("Juan", "Calle 1", "123456789", "12345678A")
cliente2 = Cliente("Pedro", "Calle 2", "987654321", "87654321B")

cuenta1 = CuentaBancaria(cliente1, "ES123456789")
cuenta2 = CuentaBancaria(cliente2, "ES987654321")

cuenta1.depositar(1000)
assert (cuenta1.transferir(cuenta2, 500) is True)

assert cuenta1.saldo == 500
assert cuenta2.saldo == 500

assert (cuenta1.transferir(cuenta2, 1000) is False)
assert cuenta1.saldo == 500    
assert cuenta2.saldo == 500
  ```


### Ejercicio 3

Crea la clase género sexo `GeneroSexo` con el atributo `genero (int)` siendo 1:masculino, 2:femenino 3:otro. <br>
Métodos: constructor, sin getters y esMasculino, esFemenino, esOtro (boolean).


Crea las siguientes clases base:
- `Direccion` tiene los atributos `calle, ciudad, codigo_postal`. Todos son de tipo String.
  Métodos: constructor, solo getters tipo @property.
- `Persona` abstracta tiene los atributos `nombre, edad, genero (GeneroSexo), direccion (Direccion)`. Todos son de tipo String.<br>
  Métodos: constructor, solo getters tipo @property.

A partir de las clases `Persona`, debemos construir:
- Clase `Empleado` que hereda de `Persona` y tiene los atributos `id_empleado (int), salario (float)`. <br>
  Métodos: constructor (todos los atributos), solo getters tipo @property.
- Clase `Cliente` que hereda de `Persona` y tiene los atributos `id_cliente (int), IdenFiscal (string)`. <br>
  Métodos: constructor (todos los atributos), solo getters tipo @property.

Además, crea una clase `Empresa` que tiene los atributos `nombre, empleados, clientes`. 
- El atributo `empleados` es una lista de objetos de la clase `Empleado`. (privado)
- El atributo `clientes` es una lista de objetos de la clase `Cliente` (privado).<br>
  Métodos: 
  - constructor (solo el nombre de la empresa)
  - Empleados (getter). Retorna la lista de empleados, pero creando una lista nueva (no retornar la lista original).
  - Clientes (getter). Retorna la lista de clientes, pero creando una lista nueva (no retornar la lista original).
  - `agregarEmpleado(empleado)`: recibe un objeto de la clase `Empleado` y lo agrega a la lista de empleados.
  - `agregarCiente(cliente)`: recibe un objeto de la clase `Cliente` y lo agrega a la lista de clientes.
  - `obtenerSalario_total()`: devuelve la suma de los salarios de todos los empleados.

Utiliza la siguiente validación para comprobar la funcionalidad:
```python
hombre = GeneroSexo(1)
mujer = GeneroSexo(2)
otro = GeneroSexo(3)

direccion1 = Direccion("Calle 1", "Ciudad 1", "12345")
direccion2 = Direccion("Calle 2", "Ciudad 2", "54321")

empleado1 = Empleado("Juan", 30, hombre, direccion1, 1, 1000)
empleado2 = Empleado("Manuela", 40, mujer, direccion2, 2, 2000)

cliente1 = Cliente("Ana", 20, mujer, direccion1, 1, "12345678A")

empresa = Empresa("Empresa 1")
empresa.agregarEmpleado(empleado1)
empresa.agregarEmpleado(empleado2)

empresa.agregarCliente(cliente1)

assert empresa.obtenerSalario_total() == 3000
assert empresa.empleados[0].nombre == "Juan"
assert empresa.empleados[1].nombre == "Manuela"
assert empresa.clientes[0].nombre == "Ana"

assert len(empresa.empleados) == 2
assert len(empresa.clientes) == 1
```
